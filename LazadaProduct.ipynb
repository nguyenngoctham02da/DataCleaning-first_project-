```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
```

```python
df1=pd.read_csv('/content/drive/MyDrive/DatasetDA/LazadaProductData.csv')
df1.head()
```

```python
#drop cột hình ảnh vì không cần thiết
df1.drop(columns=['p_image'],inplace=True)
df1.head()
```

```python
df1.isnull().sum()
```

```python
#những dòng không có rate thay thế bằng 0
df1=df1.fillna(value={'p_rate1star':0,'p_rate2star':0,'p_rate2star':0,'p_rate3star':0,'p_rate4star':0,'p_rate5star':0,'p_rating':0,'s_rating':0})
df1.isnull().sum()
```

```python
#sản phẩm có brand null thay bằng No Brand
df1=df1.fillna(value={'p_brand':'No Brand'})
df1.isnull().sum()
```

```python
df1.info()
```

```python
#chuyển p_number_reviews về int (không có đánh giá => 0, có chuyển về số)
#chuyển p_price về dạng int (bỏ đ)
#chuyển s_rating, s_response_rate, s_ship_ontime về int (bỏ %)
```

# Chuyển price về dạng số

```python
#chuyển p_price về dạng int (bỏ đ)
#tách số và ký tự đ
df1['p_price_split']=df1['p_price'].str.split(' ')
df1
```

```python
#cột p1_price là giá tiền được lấy phần tử đầu tiên trong mảng
df1['p1_price']=df1['p_price_split'].str.get(0).str.strip()
df1
```

```python
df1.info()
```

```python
#bỏ ký tự chấm trong giá tiền
df1['p1_price'] = df1['p1_price'].str.replace('.','')
df1
```

```python
df1['p1_price'] = df1['p1_price'].astype(int)
df1.head()
```

```python
df1.info()
```

#Chuyển rating về dạng số

```python
df2=df1.copy()
```

```python
#thay kí tự % bằng ký tự rỗng ở cột s_rating
df2['s_rating'] = df1['s_rating'].str.replace('%','')
df2
```

```python
#thay kí tự % bằng ký tự rỗng ở cột s_response_rate
df2['s_response_rate'] = df1['s_response_rate'].str.replace('%','')
df2.head()
```

```python
df2.info()
```

```python
#lấy ra index của những dòng có s_response_rate là không đủ thông tin
index = df2[df2['s_response_rate']=='Không đủ thông tin'].index
```

```python
#xóa những dòng có s_response_rate là không đủ thông tin
df2.drop(index,inplace=True)
```

```python
#thay kí tự % bằng ký tự rỗng ở cột s_response_rate
df2['s_ship_ontime'] = df2['s_ship_ontime'].str.replace('%','')
df2.head()
```

```python
#lấy ra index của những dòng có s_ship_ontime là không đủ thông tin và xóa chúng
index1 = df2[df2['s_ship_ontime']=='Không đủ thông tin'].index
df2.drop(index1,inplace=True)
```

```python
df2.isnull().sum()
```

```python
#những ô trống là những ô rating, thay chúng bằng 0
df2=df2.fillna(0)
df2
```

```python
#chuyển rating về dạng float
df2['s_ship_ontime']=df2['s_ship_ontime'].astype(float)
df2['s_rating']=df2['s_rating'].astype(float)
df2['s_response_rate']=df2['s_response_rate'].astype(float)
df2['p1_price']=df1['p1_price'].astype(int)
df2.info()
```

#Chuyển reviews về dạng số

```python
df3=df2.copy()
```

```python
#thay những dòng 'không có đánh giá bằng' bằng 0
df3['p_number_reviews'] = df3['p_number_reviews'].str.replace('Không có đánh giá','0')
df3.head()
```

```python
#tách giá trị ở các ô review bằng ký tự ' ', tạo thành mảng có phần tử đầu tiên là số review
df3['split_reviews']=df3['p_number_reviews'].str.split(' ')
df3.head()
```

```python
#lấy phần tử đầu tiên của mảng (là số review)
df3['number_reviews']=df3['split_reviews'].str.get(0).str.strip()
df3.head()
```

```python
#chuyển phần tử vừa lấy được về dạng số
df3['number_reviews']=df3['number_reviews'].astype(int)
df3.info()
```

```python
#drop những cột không cần thiết
df=df3.drop(['p_number_reviews','p_price','p_price_split','split_reviews'],axis=1)
df.head()
```

```python
df.info()
```

# EDA

```python
df.head()
```

```python
#tổng số lượng sản phẩm
sum_products=df['p_name'].nunique()
sum_products
```

```python
#tổng số lượng gian hàng
sum_stores=df['s_name'].nunique()
sum_stores
```

```python
#tổng số lượng thương hiệu
sum_brands=df['p_brand'].nunique()
sum_brands
```

```python
#tổng số lượng ngành hàng
sum_cates=df['p_cate'].nunique()
sum_cates
```

```python
#liệt kê những gian hàng với thương hiệu và số lượng sản phẩm tương ứng
df_store_brand=df.groupby(['s_name','p_brand'])['p_name'].count().reset_index()
df_store_brand.sort_values(by='p_name', ascending=False)
```

```python
#Những gian hàng có rating, response và ship rate cao nhất và thấp nhất
df_metrics_store=df.groupby('s_name')['s_rating','s_response_rate','s_ship_ontime'].max().reset_index()
df_metrics_store.sort_values(by=['s_rating','s_response_rate','s_ship_ontime'], ascending=False)
```

```python
#Sản phẩm lần lượt có reviews và rating cao nhất
df1_product_rating=pd.DataFrame({'Rating':df.groupby('p_name')['p_rating'].mean(),
                                 'Reviews':df.groupby('p_name')['number_reviews'].max()
    
}).reset_index()
df1_product_rating.sort_values(by=['Reviews','Rating'],ascending=False)
```

```python
#Tỷ trọng về số lượng sản phẩm mall và non-mall
df_mall_nonmall=df.groupby('p_mall')['p_name'].nunique().reset_index()
df_mall_nonmall
```

```python
plt.figure(figsize=(15,7))
plt.pie(df_mall_nonmall['p_name'],labels=df_mall_nonmall['p_mall'],autopct='%1.1f%%')
plt.legend()
```

```python
#số lượng sản phẩm theo loại
df_product_cate=df.groupby('p_cate')['p_name'].nunique().reset_index()
df_product_cate
```

```python
fig, ax=plt.subplots(nrows=2,figsize=(20,10))
ax[0].pie(df_product_cate['p_name'],labels=df_product_cate['p_cate'],autopct='%1.1f%%')
# ax[0].legend(position='right')
ax[0].set_title('Percentage of products of each category')

ax[1].bar(df_product_cate['p_cate'],df_product_cate['p_name'])
ax[1].tick_params(axis='x', rotation=30)
ax[1].set_xlabel('Category')
ax[1].set_ylabel('Number of products')
ax[1].set_title('Number of product by category')
```
